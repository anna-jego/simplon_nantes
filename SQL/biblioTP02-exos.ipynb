14.	Quelle est la durée moyenne des retards pour l’ensemble des emprunts.

La fonction « if » permet de tester une valeur et de renvoyer ce qu’on souhaite selon la vérité ou la fausseté de la valeur testée.

if(a<0, 0, a) : permet de ramener les valeurs négatives de a à 0.

-- Liste des retards

```python
pd.read_sql_query('select * from emprunter;', db)
```

```python
# Date de retour escompté

pd.read_sql_query('select nl,TIMESTAMPADD(day, dureeMax, datEmp) as retour_prevu from emprunter;', db)
```

```python
# Date de retour escompté avec le champ dateRet
pd.read_sql_query('select nl,TIMESTAMPADD(day, dureeMax, datEmp) as retour_prevu, dateRet from emprunter;', db)
```

```python
# On met la date courante pour les valeurs NULL de dateRet
pd.read_sql_query('select nl,TIMESTAMPADD(day, dureeMax, datEmp) as retour_prevu, IFNULL(dateRet, curdate()) as date_test from emprunter;', db)
```

```python
# Si retour_prevu > date_test on comptablise la ligne 
pd.read_sql_query('select *, TIMESTAMPDIFF(DAY ,IFNULL(dateRet, curdate()), TIMESTAMPADD(day, dureeMax, datEmp)) as retard from emprunter where TIMESTAMPDIFF(DAY ,IFNULL(dateRet, curdate()), TIMESTAMPADD(day, dureeMax, datEmp)) < 0;', db)
```

```python
# La moyenne
pd.read_sql_query('select AVG(TIMESTAMPDIFF(DAY ,TIMESTAMPADD(day, dureeMax, datEmp), IFNULL(dateRet, curdate()))) as moyenne_retard from emprunter where TIMESTAMPDIFF(DAY ,IFNULL(dateRet, curdate()), TIMESTAMPADD(day, dureeMax, datEmp)) < 0;', db)
```
