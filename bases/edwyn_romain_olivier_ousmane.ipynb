# **LISTES**

## definition

**Les listes sont des objets qui peuvent en contenir d’autres.**  
Ce sont donc des séquences, comme les chaînes de caractères, mais au lieu de contenir des caractères, elles peuvent contenir n’importe quel objet.


## caracteristiques

-  Objet conteneur d’autres objets

-  Pour accéder aux objets contenus, il faut connaître leur position dans la liste

-  Cette position se traduit par des entiers, appelés indices, compris entre 0 (inclus) et la taille de la liste (non incluse)

-  les listes permettent aussi de composer des matrices en plusieurs dimensions (mais je ne sais pas le faire !)


## syntaxe

###**Création de Liste**


```
liste1 = [1,2,3]                  # peut contenir des integers ou floats
liste2 = ["blue","red","green"]   # peut contenir des strings ou d'autres objets
liste3 = [liste1,liste2]          # peut contenir une autre liste
liste4 = []
```

###**Accès aux elements**

```
print(liste1)      # affiche la liste
print(liste1[0])   # affiche le premier element de la liste
print(liste1[-1])  # affiche le dernier element de la liste
liste1[1] = 50     # modifie le second element de la liste
print(liste1)      # affiche la liste (après modification)
```

###**Ajout d’éléments**

- append

```
liste2.append("purple")    # ajoute un element à la fin de la liste
print(liste2)
```

-  insert

```
liste2.insert(2, "yellow")          # insert un element en troisieme position dans la liste
print(liste2)
```

-  extend

```
liste4 = [4,5]
liste5 = [6,7,8]
liste4.extend(liste5)      # insertion de liste5 à la fin de liste4
print(liste4)
```

###**suppression d'éléments**

-  del

```
liste6 = [1,2,3,4,5,6]
del liste6[4]             # suppression du 5eme élément de la liste
print(liste6)
```

-  remove

```
liste7 = [789,78,456,23,456]
liste7.remove(456)             # suppression du premier élément 456 de la liste
print(liste7)
```

###**parcourir la liste**

-  while

```
liste8 = [1,2,3,4]
i = 0
while i < len(liste8):    # tant que i est plus petit que le nombre d'element dans la liste (len pour lenght)
  print(liste8[i])        # affiche l'element ciblé
  i += 1                  # incrementation pour passer à l'élément suivant

```

-  for (recommandé, apparrement, c'est internet qui l'a dit)




```
for element in liste8:      #element prend la valeur des differents elements
  print(element)            #on print les differents elements
```

### action utilisable

x in s ....................... Vrai si l’élément x est dans la liste s.  
x not in s ................... Vrai si l’élément x ne figure pas dans la liste s.  
s1 + s2 .....................	Concaténer deux listes s1 et s2.  
s * n, n * s ................ Crée une liste composée de n copies concaténées de s.  
s[i] ........................ Elément d’indice i dans la liste s.  
s[i : j] .................... Tranche de la liste s dès l’indice i à l’indice j -1.  
len(s) ...................... Le nombre d’éléments dans la liste s.  
min(s) ...................... Le plus petit élément dans la liste s.  
max(s) ...................... Le plus grand élément dans la liste s.  
sum(s) ...................... Somme de tous les éléments de s.  
Boucle for .................. Traverse les éléments de la liste s de gauche à droite.  
<, <=, >, >=, =, != ......... Compare deux listes  

# **DICTIONNAIRES**

## definition

- objet conteneur associant des clés à des valeurs
- exemple :
  - mon_dictionnaire = {"couleur": "rouge", "taille": 39, "marque": "Levis"}
- si la clé n’existe pas dans le dictionnaire, une exception de type **KeyError** sera levée
- un dictionnaire ne peut pas contenir deux clés identiques, la 2nde valeur écrasant la première pour un même nom de clé
- en revanche rien n’empêche d’avoir deux valeurs identiques dans deux clés différentes


## caracteristiques

- pas d'ordre précis
- associe chaque objet contenu à une clé (généralement une chaîne de caractères)

- **liste vs dictionnaire**
  - aucune structure ordonnée
  - pour accéder aux objets contenus, pas nécessairement des  indices mais des clés qui peuvent être des types distincts

## syntaxe

dictionnaire = { cle1:valeur1, cle2:valeur2, cleN:valeurN }

### création

* avec la méthode **dict()**

```
mon_dictionnaire = dict() # creation d'un dictionnaire vide

print(type(mon_dictionnaire))
print(mon_dictionnaire)
```

* avec les accolades   **{}**

```
mon_dictionnaire = {"couleur": "rouge", "taille": 39, "marque": "Levis"}
print(mon_dictionnaire)

print(type(mon_dictionnaire))
print(mon_dictionnaire)
```

### ajout

```
# creation d'un dictionnaire vide
mon_dictionnaire = {}
print(mon_dictionnaire)

# ajout d'élements
mon_dictionnaire["couleur"] = "rouge"
mon_dictionnaire["taille"] = 39
print(mon_dictionnaire)
```

### suppression 

* avec le mot clé : **del**

```
mon_dictionnaire = {"couleur": "rouge", "taille": 39, "marque": "Levis"}
print(mon_dictionnaire)

del mon_dictionnaire['taille']
print(mon_dictionnaire)
```

* avec le mot clé : **pop**

```
mon_dictionnaire = {"couleur": "rouge", "taille": 39, "marque": "Levis"}
print(mon_dictionnaire)

mon_dictionnaire.pop('taille')
print(mon_dictionnaire)
```

### parcourir un dictionnaire

#### **parcours des clefs**

* avec une boucle **for**

```
# Il est possible de parcourir les clés d’un dictionnaire de la même manière qu’une liste

mon_dictionnaire = {"pommes": 5, "melons": 7, "poires": 9}

for key in mon_dictionnaire:
  print(key)
```

* avec la méthode **keys()** de la classe **dict**

```
mon_dictionnaire = {"pommes": 5, "melons": 7, "poires": 9}

print(mon_dictionnaire.keys())

for key in mon_dictionnaire.keys():
  print(key)
```

#### **parcours des valeurs**

* avec la méthode **values()**

```
mon_dictionnaire = {"pommes": 5, "melons": 7, "poires": 9}

print(mon_dictionnaire.values())

for key in mon_dictionnaire.values():
  print(key)
```

#### **parcours des clés et des valeurs simultanément**

* avec la méthode **items()**

```
mon_dictionnaire = {"pommes": 5, "melons": 7, "poires": 9}

print(mon_dictionnaire.items())

for key, value in mon_dictionnaire.items():
  print(key, value)
```

# **TUPLES**





## definition

Les tuples sont des collections d’éléments, mais leurs éléments sont fixés; càd, une fois le tuple est créé, vous ne pouvez pas ajouter de nouveaux éléments, supprimer des éléments, remplacer des éléments, ou réorganiser les éléments du tuple.

Si le contenu d’une liste dans votre programme/fonction ne devrait pas changer, vous pouvez utiliser un tuple pour éviter que des éléments soient ajoutés, supprimés ou remplacés accidentellement.

Un tuple est très semblable à une liste, sauf que ses éléments sont fixés. En outre, les tuples sont plus efficaces que les listes dues aux implémentations de Python.

## caracteristique

Type Immuable par defaut

En outre, si un tuple X contient un élément y de type muable. On peut changer la valeur de Y.

## syntaxe









 


####**Déclaration d'un tuple:**

```
t1 = () # Crée un tuple vide

t2 = (1, 3, 5) # Crée un tuple constitué de trois éléments

t3 = (2,) # Crée un tuple constitué d’un seul élément

t3 = tuple([2 * x for x in range(1, 5)])# Crée un tuple à partir d’une liste
```

####**Lors de la déclaration d'un tuple avec une chaîne de caratères, chaque caractère deviens un éléments du tuple, soit:**

```
t4 = tuple("abac") # t4 est ('a', 'b', 'a', 'c')
```

####**La liste ci-dessous résume les opérations utilisées avec les tuples:**

```
x in s : Vrai si l’élément x est dans le tuple s.

x not in s : Vrai si l’élément x ne figure pas dans le tuple s.

s1 + s2 : Concaténer deux tuples s1 et s2.

s * n, n * s : Crée un tuple composé de n copies concaténées de s.

s[i] : Elément d’indice i dans le tuple s.

s[i : j] : Tranche du tuple s dès l’indice i à l’indice j – 1.

len(s) : Le nombre d’éléments dans le tuple s.

min(s) : Le plus petit élément dans le tuple s.

max(s) : Le plus grand élément dans le tuple s.

sum(s) : Somme de tous les éléments de s.

Boucle for : Traverse les éléments du tuple s de gauche à droite.

<, <=, >, >=, =, != : Compare deux tuples
```
